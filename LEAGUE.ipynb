```python
# Line graph comparing 27 different countries in terms of average compulsory schooling years over the years
import matplotlib.pyplot as plt
from scipy import stats

x1 = life_expectancy['Average schooling years'].iloc[0:27]
y1 = life_expectancy['Average Life expectancy'].iloc[0:27]

slope, intercept, r, p, std_err = stats.linregress(x1, y1)

def myfunc(x1):
  return slope * x1 + intercept

mymodel = list(map(myfunc, x1))
plt.title('Finding relationship between schooling years and life expectancy')
plt.xlabel('Average schooling years')
plt.ylabel('Average Life expectancy')
plt.scatter(x1, y1,color='r')
plt.plot(x1, mymodel,color='royalblue')
plt.show()
```
